```python
# 選一個示範使用
y_col = ['ARO1-DCS-PX 回收率']
```

```python
def get_interval(df,col):
    interval = []
    for t,i in enumerate(df[col]):
        if str(i) != 'nan':
            interval.append(t)
    interval = [ i-j for i,j in zip(interval[1:],interval[:-1])]
    return max(interval,key=interval.count)
get_interval(df_y,y_col[0])
```

```python
df_xy = df_x[s_col+a_col].join(df_y[y_col])
df_xy[s_col+a_col] = df_xy[s_col+a_col].fillna(method='bfill')
df_xy
```

```python
from tqdm import tqdm
for i in tqdm(df_xy.columns):
    df_xy[i] = pd.to_numeric(df_xy[i],errors='coerce')
```

```python
from sklearn.preprocessing import MinMaxScaler

mm_a = MinMaxScaler()
mm_s = MinMaxScaler()
mm_y = MinMaxScaler()

df_xy[a_col] = mm_a.fit_transform(df_xy[a_col])
df_xy[s_col] = mm_s.fit_transform(df_xy[s_col])
df_xy[y_col] = mm_y.fit_transform(df_xy[y_col])
df_xy
```

# 建立資料集 for 建模使用

```python
from tqdm import tqdm_notebook as tqdm
def build_dataset(df,n=None):
    data = {
        'state':[],
        'action':[],
        'value':[],
        'idx':[],
        }
    
    for i in tqdm(range(0,len(df)-n,n)):
        curr = df.iloc[i:i+n] #一小段資料
        cond1 = np.isnan(curr[y_col].values[-1]).sum() == 0
        cond2 = np.isnan(curr[s_col+a_col].values).sum() == 0
        if cond1 & cond2: #確保這段資料無缺失值
            data['state'].append(curr[s_col].values) # 這段時間所有狀態
            data['action'].append(curr[a_col].values) # 這段時間所有操作
            data['value'].append(curr[y_col].values[-1]) # 這段時間最後一個時間點的產量
            data['idx'].append(curr.index[-1]) #這段時間最後 出料對應的時間點
    
    # reshape
    data['state'] = np.array(data['state'])
    data['action'] = np.array(data['action'])
    data['value'] = np.array(data['value'])
    
    # store mm_scaler
    data['mm_state'] = mm_s
    data['mm_action'] = mm_a
    data['mm_value'] = mm_y
    
    # store col_name
    data['state_col'] = s_col
    data['action_col'] = a_col
    data['value_col'] = y_col
    
    return data
```

```python
import numpy as np
data = build_dataset(df_xy.iloc[df_xy[y_col].dropna(axis=0).index[0]+1:],n=get_interval(df_y,y_col[0]))
```
