# 建立資料集 for 建模使用

```python
for j,i in enumerate(df_xy[y_col].values.reshape(-1)):
    if str(i) != 'nan':
        print(j,i)
```

```python
from tqdm import tqdm_notebook as tqdm
def build_dataset(df,n=8): 
    data = {
        'x':[],
        'y':[],
        'idx':[],
        }
    
    for i in tqdm(range(0,len(df)-n,n)):
        curr = df.iloc[i:i+n] #一小段資料
        if np.isnan(curr[y_col].values[-1]).sum() == 0: #確保這段資料無缺失值
            data['x'].append(curr[x_col].values)
            data['y'].append(curr[y_col].values[-1])
            data['idx'].append(curr.index[-1]) #這段時間最後 出料對應的時間點
    
    # reshape
    data['x'] = np.array(data['x'])
    data['y'] = np.array(data['y'])
    
    # store mm_scaler
    data['mm_x'] = mm_x
    data['mm_y'] = mm_y
    
    # store col_name
    data['x_col'] = x_col
    data['y_col'] = y_col
    
    return data
```

```python
import numpy as np
data = build_dataset(df_xy.iloc[48343+1:],n=8)
```
