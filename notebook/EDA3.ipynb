```python
import pandas as pd
path = '../data/R911R912 _明志蔡教授.xlsx'

xls = pd.ExcelFile(path)
sheets = xls.sheet_names
display(sheets)
df = pd.read_excel(path,sheet_name=sheets[-1])
for i in df.columns:
    df[i] = pd.to_numeric(df[i],errors='coerce') 
df = df.dropna(axis=1,how='all').set_index('Unnamed: 2')
df.index.name = 'time'
df.index = pd.to_datetime(df.index)
df.head()
```

```python
import pandas as pd
path = '../data/R911R912 _ARO1.xlsx'

xls = pd.ExcelFile(path)
sheets = xls.sheet_names
display(sheets)
df2 = pd.read_excel(path,sheet_name=sheets[-2])
for i in df2.columns:
    df2[i] = pd.to_numeric(df2[i],errors='coerce') 
df2 = df2.dropna(axis=1,how='all').set_index('Unnamed: 1')
df2.index.name = 'time'
df2.index = pd.to_datetime(df2.index)
df2.head()
```

```python
import seaborn as sb

a = df
b = df2

# L2
L2_a = a.columns[a.columns.str.contains('L2')]
L2_b = b.columns[b.columns.str.contains('L2')]
a[L2_a[0]] = pd.to_numeric(a[L2_a].values.reshape(-1),errors='coerce') 
b[L2_b[0]] = pd.to_numeric(b[L2_b].values.reshape(-1),errors='coerce') 
sb.kdeplot(a[L2_a[0]])
sb.kdeplot(b[L2_b[0]])
```
