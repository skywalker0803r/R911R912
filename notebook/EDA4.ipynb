```python
import pandas as pd
path = '../data/R911R912 _明志蔡教授.xlsx'

xls = pd.ExcelFile(path)
sheets = xls.sheet_names
display(sheets)
df = pd.read_excel(path,sheet_name=sheets[-1])
for i in df.columns:
    df[i] = pd.to_numeric(df[i],errors='coerce') 
df = df.dropna(axis=1,how='all').set_index('Unnamed: 2')
df.index.name = 'time'
df.index = pd.to_datetime(df.index)
print(df.shape)
df.head()
```

```python
df.columns
```

```python
x_col = df.columns[df.columns.str.contains('_L')].tolist()+df.columns[df.columns.str.contains('FA')].tolist()
```
